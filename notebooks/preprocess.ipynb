```python
def handle_missing_data(df):
    # Fill categorical missing values with "Unknown"
    categorical_cols = df.select_dtypes(include=["object"]).columns
    df[categorical_cols] = df[categorical_cols].fillna("Unknown")

    # Fill numerical missing values with median
    numerical_cols = df.select_dtypes(include=["number"]).columns
    imputer = SimpleImputer(strategy="median")
    df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

    return df
```

```python
def feature_engineering(df_rents):
    # Add contract duration in days
    df_rents["contract_duration"] = (
        pd.to_datetime(df_rents["contract_end_date"]) - pd.to_datetime(df_rents["contract_start_date"])
    ).dt.days

 
    return df_rents
```

```python
from sklearn.preprocessing import OneHotEncoder

def encode_categorical_with_frequency(df):
    categorical_cols = df.select_dtypes(include=["object"]).columns
    for col in categorical_cols:
        freq_map = df[col].value_counts().to_dict()
        df[col] = df[col].map(freq_map)
    return df

def encode_categorical(df):
    categorical_cols = df.select_dtypes(include=["object"]).columns

    # OneHotEncode categorical columns
    encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    encoded_data = pd.DataFrame(
        encoder.fit_transform(df[categorical_cols]),
        columns=encoder.get_feature_names_out(categorical_cols),
    )

    # Drop original categorical columns and merge encoded
    df = df.drop(categorical_cols, axis=1)
    df = pd.concat([df.reset_index(drop=True), encoded_data.reset_index(drop=True)], axis=1)

    return df
```

```python
def scale_numerical(df):
    numerical_cols = df.select_dtypes(include=["number"]).columns
    scaler = StandardScaler()
    df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
    return df
```

```python
def preprocess_data():
    rents = load_data()

    # Handle missing data
    rents = handle_missing_data(rents)
 
    # Perform feature engineering
    rents = feature_engineering(rents)

    # Encode categorical variables
    rents = encode_categorical_with_frequency(rents)
 
    # Scale numerical features
    rents = scale_numerical(rents)
 
    return rents
```

```python
rents, transactions = preprocess_data()
```
